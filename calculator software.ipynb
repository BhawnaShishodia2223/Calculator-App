```python
from tkinter import *

root=Tk()
root.title("Calculator")
root.resizable(0,0)
root.geometry("250x300")

x=0
value=""
operator=""

def press1():
    global value
    value=value+"1"
    data.set(value)

def press2():
    global value
    value=value+"2"
    data.set(value)

def press3():
    global value
    value=value+"3"
    data.set(value)

def press4():
    global value
    value=value+"4"
    data.set(value)
    
def press5():
    global value
    value=value+"5"
    data.set(value)
    
def press6():
    global value
    value=value+"6"
    data.set(value)
    
def press7():
    global value
    value=value+"7"
    data.set(value)
    
def press8():
    global value
    value=value+"8"
    data.set(value)
    
def press9():
    global value
    value=value+"9"
    data.set(value)
    
def press0():
    global value
    value=value+"0"
    data.set(value)

def pressPlus():
    global value,operator
    global x
    x=int(value)
    operator="+"
    value=value+"+"
    data.set(value)
    
def pressMinus():
    global value,operator
    global x
    operator="-"
    x=int(value)
    value=value+"-"
    data.set(value)

def pressMult():
    global value,operator
    global x
    operator="*"
    x=int(value)
    value=value+"*"
    data.set(value)

def pressDiv():
    global value, operator
    global x
    x=int(value)
    operator="/"
    value=value+"/"
    data.set(value)

def pressClear():
    global value,operator
    global x
    value=""
    operator=""
    x=0
    data.set(value)

    
def result():
    global value,operator
    global x
    
    if operator=="+" :
        y=int((value.split("+")[1]))
        z=x+y
        value=str(z)
        data.set(value)
    
    
    if operator=="-" :
        y=int(value.split("-")[1])
        z=x-y
        value=str(z)
        data.set(value)

    if operator=="*":
        
        y=int(value.split("*")[1])
        z=x*y
        value=str(z)
        data.set(value)
    
    
    if operator=="/":
        
        y=int(value.split("/")[1])
        if y==0:
            error="Division Error"
            data.set(error)   
        else :
            z=int(x/y)
            value=str(z)
            data.set(value)
    
    
    
    
    
    
data=StringVar()

#now we are making the frames
row_1=Label(root,text="label",bg="#ffffff",anchor=SE,font=("Geneva",15),textvariable=data).pack(expand=True,fill="both") #row1 
  
row_2=Frame(root)
row_2.pack(expand=True,fill="both") #row2

row_3=Frame(root)
row_3.pack(expand=True,fill="both") #row3

row_4=Frame(root)
row_4.pack(expand=True,fill="both") #row4

row_5=Frame(root)
row_5.pack(expand=True,fill="both") #row5



#buttons section

#button1
btn_1=Button(row_2,text="1",border=0,font=("Geneva",15),command=press1).pack(expand=True,fill="both",side=LEFT)
#button2
btn_2=Button(row_2,text="2",border=0,font=("Geneva",15),command=press2).pack(expand=True,fill="both",side=LEFT)
#button3
btn_3=Button(row_2,text="3",border=0,font=("Geneva",15),command=press3).pack(expand=True,fill="both",side=LEFT)
#button-plus
btn_plus=Button(row_2,text="+",border=0,font=("Geneva",15),command=pressPlus).pack(expand=True,fill="both",side=LEFT)


#button4
btn_4=Button(row_3,text="4",border=0,font=("Geneva",15),command=press4).pack(expand=True,fill="both",side=LEFT)
#button5
btn_5=Button(row_3,text="5",border=0,font=("Geneva",15),command=press5).pack(expand=True,fill="both",side=LEFT)
#button6
btn_6=Button(row_3,text="6",border=0,font=("Geneva",15),command=press6).pack(expand=True,fill="both",side=LEFT)
#button-minus
btn_minus=Button(row_3,text="-",border=0,font=("Geneva",15),command=pressMinus).pack(expand=True,fill="both",side=LEFT) 


#button7
btn_7=Button(row_4,text="7",border=0,font=("Geneva",15),command=press7).pack(expand=True,fill="both",side=LEFT)
#button8
btn_8=Button(row_4,text="8",border=0,font=("Geneva",15),command=press8).pack(expand=True,fill="both",side=LEFT)
#button9
btn_9=Button(row_4,text="9",border=0,font=("Geneva",15),command=press9).pack(expand=True,fill="both",side=LEFT)
#button-mult
btn_mult=Button(row_4,text="*",border=0,font=("Geneva",15),command=pressMult).pack(expand=True,fill="both",side=LEFT)



#button-clear
btn_cls=Button(row_5,text="Clear",border=0,font=("Geneva",15),command=pressClear).pack(expand=True,fill="both",side=LEFT)
#button0
btn_0=Button(row_5,text="0",border=0,font=("Geneva",15),command=press0).pack(expand=True,fill="both",side=LEFT)
#button-equal
btn_equal=Button(row_5,text="=",border=0,font=("Geneva",15),command=result).pack(expand=True,fill="both",side=LEFT)
#button-div
btn_div=Button(row_5,text="/",border=0,font=("Geneva",15),command=pressDiv).pack(expand=True,fill="both",side=LEFT)



mainloop()
```
